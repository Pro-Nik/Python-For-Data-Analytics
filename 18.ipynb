Bar plot in seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

Another way to show bargraph

```python
sns.barplot(x = "island" , y ="bill_length_mm", data= df1)
plt.show()
```

```python
df1
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data= df1, hue="sex")
plt.show()
```

Order Parameter

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x= "island", y = "bill_length_mm", data= df1, hue= "sex", order = order1)
plt.show()
```

hue_order

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x= "island", y = "bill_length_mm", data= df1, hue= "sex", order = order1, hue_order= ['Female','Male'])
plt.show()
```

```python
order1 = ["Dream","Trogersen","Biscoe"]
sns.barplot(x= "island",y ="bill_length_mm", data= df1, color="green")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", hue = "sex", data= df1, palette="Accent")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", hue = "sex", data= df1, palette="prism")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", hue = "sex", data= df1, palette="gist_earth")
plt.show()
```

Saturation

```python
sns.barplot(x = "island", y = "bill_length_mm", hue = "sex", data= df1, palette="prism", saturation= 0.2)
plt.show()
```

Histogram in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

Bins

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], color= "green")
plt.show()
```

kde

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde= True, color= "green")
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde= False, color= "pink")
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde= True, color= "pink")
plt.show()
```
