Scatter Plot

```python
x = [1,2,3,4,5,6]
y = [2,6,7,4,5,6]

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Number")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,6,7,4,5,6]

c = ['r','g','y','b','g','r']
plt.scatter(x,y, color = c)
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Number")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,6,7,4,5,6]

c = ['r','g','y','b','g','r']
plt.scatter(x,y, color = c , s = 100)
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Number")
plt.show()
```

Pie-Chart

```python
x = [10,20,30,40]
y =['English','Hindi','Maths','Science']

plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y =['English','Hindi','Maths','Science']

plt.pie(x, labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y =['English','Hindi','Maths','Science']

c = ["yellow","blue","green","red"]
plt.pie(x, labels = y , colors = c)
plt.show()
```

```python
x = [10,20,30,40]
y =['English','Hindi','Maths','Science']

c = ["yellow","blue","green","red"]
plt.pie(x, labels = y , colors = c)
plt.legend()
plt.show()
```
